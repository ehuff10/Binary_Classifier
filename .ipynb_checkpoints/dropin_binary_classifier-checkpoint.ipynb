## Utility functions

These helpers keep the rest of the notebook clean and reusable.

```python
# Utility helpers used throughout the notebook
# Keeping these here makes the rest of the notebook easier to read.

def train_test_split_np(X, y, test_size=0.2, seed=42):
    """Simple NumPy train/test split."""
    X = np.asarray(X, dtype=float)
    y = np.asarray(y).reshape(-1)

    n = X.shape[0]
    rng = np.random.default_rng(seed)
    idx = rng.permutation(n)

    test_n = int(n * test_size)
    test_idx = idx[:test_n]
    train_idx = idx[test_n:]

    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]


def confusion_counts(y_true, y_pred):
    """Return TP, TN, FP, FN counts for binary labels in {0,1}."""
    y_true = np.asarray(y_true).reshape(-1)
    y_pred = np.asarray(y_pred).reshape(-1)

    tp = int(np.sum((y_true == 1) & (y_pred == 1)))
    tn = int(np.sum((y_true == 0) & (y_pred == 0)))
    fp = int(np.sum((y_true == 0) & (y_pred == 1)))
    fn = int(np.sum((y_true == 1) & (y_pred == 0)))
    return tp, tn, fp, fn


def classification_metrics(tp, tn, fp, fn):
    """Compute precision, recall, F1, accuracy."""
    precision = tp / (tp + fp + 1e-12)
    recall = tp / (tp + fn + 1e-12)
    f1 = 2 * precision * recall / (precision + recall + 1e-12)
    accuracy = (tp + tn) / (tp + tn + fp + fn + 1e-12)
    return float(precision), float(recall), float(f1), float(accuracy)


def brier_score(y_true, probs):
    """Calibration-style score: lower is better."""
    y_true = np.asarray(y_true).reshape(-1)
    probs = np.asarray(probs).reshape(-1)
    return float(np.mean((probs - y_true) ** 2))


def log_loss(y_true, probs, eps=1e-12):
    """Binary log loss (cross-entropy)."""
    y_true = np.asarray(y_true).reshape(-1)
    probs = np.asarray(probs).reshape(-1)
    probs = np.clip(probs, eps, 1 - eps)
    return float(-np.mean(y_true * np.log(probs) + (1 - y_true) * np.log(1 - probs)))
```

## Standardization

Neural nets train better when features are on a similar scale.

```python
# Standardization helpers
# Neural nets are much easier to train when features are centered/scaled.

def fit_standardizer(X_train):
    mu = X_train.mean(axis=0)
    sigma = X_train.std(axis=0) + 1e-12
    return mu, sigma


def transform_standardizer(X, mu, sigma):
    return (X - mu) / sigma


# Fit on training only, then apply to train + test
mu, sigma = fit_standardizer(X_train)
X_train_s = transform_standardizer(X_train, mu, sigma)
X_test_s = transform_standardizer(X_test, mu, sigma)

print("Scaled train mean (first 5):", np.mean(X_train_s, axis=0)[:5])
print("Scaled train std  (first 5):", np.std(X_train_s, axis=0)[:5])
```

## Neural network components

We’ll implement a small 2-layer network:

Input → ReLU → Sigmoid

```python
# Small math helpers for the network
# We clip in a couple places to avoid overflow in exp/log.

def sigmoid(z):
    z = np.clip(z, -50, 50)
    return 1 / (1 + np.exp(-z))


def relu(z):
    return np.maximum(0, z)


def drelu(z):
    # Derivative of ReLU
    return (z > 0).astype(float)


def bce_loss(y_true_row, y_pred_row, eps=1e-12):
    # Binary cross entropy
    y_pred_row = np.clip(y_pred_row, eps, 1 - eps)
    return float(-np.mean(
        y_true_row * np.log(y_pred_row) + (1 - y_true_row) * np.log(1 - y_pred_row)
    ))
```

## Drop‑in model class

This model expects:
- `X` as `(m, n_features)`
- `y` as `(m,)` with labels 0/1

It provides:
- `fit`
- `predict_proba`
- `predict`
- `score`
- `save` / `load`

```python
class DropInBinaryNN:
    def __init__(self, n_h=32, lr=0.01, epochs=2000, clip_value=1.0, seed=42):
        self.n_h = int(n_h)
        self.lr = float(lr)
        self.epochs = int(epochs)
        self.clip_value = float(clip_value)
        self.seed = int(seed)

        self.params_ = None
        self.losses_ = []
        self.is_fitted_ = False
        self.n_features_ = None

    def _init_params(self, n_x):
        rng = np.random.default_rng(self.seed)
        W1 = rng.standard_normal((self.n_h, n_x)) * np.sqrt(2 / n_x)  # He init
        b1 = np.zeros((self.n_h, 1))
        W2 = rng.standard_normal((1, self.n_h)) * np.sqrt(1 / self.n_h)
        b2 = np.zeros((1, 1))
        self.params_ = {"W1": W1, "b1": b1, "W2": W2, "b2": b2}

    def _forward(self, X):
        # X: (m, n_features)
        X_T = X.T  # (n_features, m)
        W1, b1 = self.params_["W1"], self.params_["b1"]
        W2, b2 = self.params_["W2"], self.params_["b2"]

        Z1 = W1 @ X_T + b1
        A1 = relu(Z1)

        Z2 = W2 @ A1 + b2
        A2 = sigmoid(Z2)

        cache = {"X_T": X_T, "Z1": Z1, "A1": A1, "A2": A2}
        return A2, cache

    def _backward(self, y_row, cache):
        # y_row: (1, m)
        m = y_row.shape[1]
        W2 = self.params_["W2"]

        X_T = cache["X_T"]
        Z1 = cache["Z1"]
        A1 = cache["A1"]
        A2 = cache["A2"]

        dZ2 = A2 - y_row
        dW2 = (1/m) * (dZ2 @ A1.T)
        db2 = (1/m) * np.sum(dZ2, axis=1, keepdims=True)

        dA1 = W2.T @ dZ2
        dZ1 = dA1 * drelu(Z1)
        dW1 = (1/m) * (dZ1 @ X_T.T)
        db1 = (1/m) * np.sum(dZ1, axis=1, keepdims=True)

        return {"dW1": dW1, "db1": db1, "dW2": dW2, "db2": db2}

    def _update(self, grads):
        for k in grads:
            grads[k] = np.clip(grads[k], -self.clip_value, self.clip_value)

        self.params_["W1"] -= self.lr * grads["dW1"]
        self.params_["b1"] -= self.lr * grads["db1"]
        self.params_["W2"] -= self.lr * grads["dW2"]
        self.params_["b2"] -= self.lr * grads["db2"]

    def fit(self, X, y, print_every=200):
        X = np.asarray(X, dtype=float)
        y = np.asarray(y).reshape(-1)

        m, n_x = X.shape
        self.n_features_ = n_x

        self._init_params(n_x)
        self.losses_ = []

        y_row = y.reshape(1, -1)

        for epoch in range(1, self.epochs + 1):
            probs, cache = self._forward(X)
            loss = bce_loss(y_row, probs)

            grads = self._backward(y_row, cache)
            self._update(grads)

            self.losses_.append(loss)

            if print_every and epoch % print_every == 0:
                print(f"Epoch {epoch:4d} | Loss: {loss:.4f}")

        self.is_fitted_ = True
        return self

    def predict_proba(self, X):
        if not self.is_fitted_:
            raise ValueError("Model not fitted. Call fit() first.")

        X = np.asarray(X, dtype=float)
        if X.shape[1] != self.n_features_:
            raise ValueError(f"Expected {self.n_features_} features, got {X.shape[1]}.")

        probs, _ = self._forward(X)
        return probs.reshape(-1)

    def predict(self, X, threshold=0.5):
        probs = self.predict_proba(X)
        return (probs >= threshold).astype(int)

    def score(self, X, y, threshold=0.5):
        y = np.asarray(y).reshape(-1)
        preds = self.predict(X, threshold=threshold)
        return float(np.mean(preds == y))

    def save(self, filepath="dropin_nn_weights.npz"):
        if not self.is_fitted_:
            raise ValueError("Train before saving.")
        np.savez(
            filepath,
            W1=self.params_["W1"], b1=self.params_["b1"],
            W2=self.params_["W2"], b2=self.params_["b2"],
            n_h=self.n_h, lr=self.lr, epochs=self.epochs,
            clip_value=self.clip_value, seed=self.seed,
            n_features=self.n_features_
        )

    @classmethod
    def load(cls, filepath="dropin_nn_weights.npz"):
        data = np.load(filepath, allow_pickle=True)
        model = cls(
            n_h=int(data["n_h"]),
            lr=float(data["lr"]),
            epochs=int(data["epochs"]),
            clip_value=float(data["clip_value"]),
            seed=int(data["seed"])
        )
        model.params_ = {"W1": data["W1"], "b1": data["b1"], "W2": data["W2"], "b2": data["b2"]}
        model.n_features_ = int(data["n_features"])
        model.is_fitted_ = True
        model.losses_ = []
        return model
```

## Full tool wrapper (scaling + model)

This wrapper stores the standardization parameters so you can save/load a single artifact and use it anywhere.

```python
class BinaryClassifierTool:
    def __init__(self, n_h=32, lr=0.01, epochs=2000, clip_value=1.0, seed=42):
        self.model = DropInBinaryNN(n_h=n_h, lr=lr, epochs=epochs, clip_value=clip_value, seed=seed)
        self.mu_ = None
        self.sigma_ = None
        self.is_fitted_ = False

    def fit(self, X_train, y_train, print_every=200):
        X_train = np.asarray(X_train, dtype=float)
        y_train = np.asarray(y_train).reshape(-1)

        self.mu_, self.sigma_ = fit_standardizer(X_train)
        Xs = transform_standardizer(X_train, self.mu_, self.sigma_)

        self.model.fit(Xs, y_train, print_every=print_every)
        self.is_fitted_ = True
        return self

    def predict_proba(self, X):
        if not self.is_fitted_:
            raise ValueError("Tool not fitted. Call fit() first.")
        X = np.asarray(X, dtype=float)
        Xs = transform_standardizer(X, self.mu_, self.sigma_)
        return self.model.predict_proba(Xs)

    def predict(self, X, threshold=0.5):
        probs = self.predict_proba(X)
        return (probs >= threshold).astype(int)

    def score(self, X, y, threshold=0.5):
        y = np.asarray(y).reshape(-1)
        preds = self.predict(X, threshold=threshold)
        return float(np.mean(preds == y))

    def save(self, filepath="binary_tool.npz"):
        if not self.is_fitted_:
            raise ValueError("Train before saving.")
        np.savez(
            filepath,
            mu=self.mu_,
            sigma=self.sigma_,
            W1=self.model.params_["W1"], b1=self.model.params_["b1"],
            W2=self.model.params_["W2"], b2=self.model.params_["b2"],
            n_h=self.model.n_h, lr=self.model.lr, epochs=self.model.epochs,
            clip_value=self.model.clip_value, seed=self.model.seed,
            n_features=self.model.n_features_
        )

    @classmethod
    def load(cls, filepath="binary_tool.npz"):
        data = np.load(filepath, allow_pickle=True)
        tool = cls(
            n_h=int(data["n_h"]),
            lr=float(data["lr"]),
            epochs=int(data["epochs"]),
            clip_value=float(data["clip_value"]),
            seed=int(data["seed"])
        )
        tool.mu_ = data["mu"]
        tool.sigma_ = data["sigma"]

        tool.model.params_ = {"W1": data["W1"], "b1": data["b1"], "W2": data["W2"], "b2": data["b2"]}
        tool.model.n_features_ = int(data["n_features"])
        tool.model.is_fitted_ = True
        tool.is_fitted_ = True
        return tool
```

### Threshold sweep

See how precision/recall/F1 change as you move the threshold.

```python
thresholds = np.linspace(0.05, 0.95, 19)
precisions, recalls, f1s, accs = [], [], [], []

for t in thresholds:
    preds = (probs_test >= t).astype(int)
    tp, tn, fp, fn = confusion_counts(y_test, preds)
    p, r, f, a = classification_metrics(tp, tn, fp, fn)
    precisions.append(p)
    recalls.append(r)
    f1s.append(f)
    accs.append(a)

plt.figure()
plt.plot(thresholds, precisions, label="Precision")
plt.plot(thresholds, recalls, label="Recall")
plt.plot(thresholds, f1s, label="F1")
plt.plot(thresholds, accs, label="Accuracy")
plt.title("Metrics vs Threshold")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.legend()
plt.show()
```

### Pick a threshold automatically

```python
def best_threshold(y_true, probs, metric="f1"):
    y_true = np.asarray(y_true).reshape(-1)
    probs = np.asarray(probs).reshape(-1)

    best_t, best_s = 0.5, -1

    for t in np.linspace(0.05, 0.95, 181):
        preds = (probs >= t).astype(int)
        tp, tn, fp, fn = confusion_counts(y_true, preds)
        precision, recall, f1, acc = classification_metrics(tp, tn, fp, fn)

        score = {"precision": precision, "recall": recall, "f1": f1, "acc": acc}[metric]
        if score > best_s:
            best_s = score
            best_t = float(t)

    return best_t, float(best_s)

best_t_f1, best_f1 = best_threshold(y_test, probs_test, metric="f1")
best_t_rec, best_rec = best_threshold(y_test, probs_test, metric="recall")

print("Best threshold for F1    :", best_t_f1, "| F1:", best_f1)
print("Best threshold for Recall:", best_t_rec, "| Recall:", best_rec)
```

### ROC curve (manual)

```python
def roc_curve_manual(y_true, probs):
    y_true = np.asarray(y_true).reshape(-1)
    probs = np.asarray(probs).reshape(-1)

    thresholds = np.r_[1.01, np.sort(np.unique(probs))[::-1], -0.01]

    tprs = []
    fprs = []

    P = np.sum(y_true == 1)
    N = np.sum(y_true == 0)

    for t in thresholds:
        preds = (probs >= t).astype(int)
        tp, tn, fp, fn = confusion_counts(y_true, preds)

        tpr = tp / (P + 1e-12)
        fpr = fp / (N + 1e-12)

        tprs.append(tpr)
        fprs.append(fpr)

    return np.array(fprs), np.array(tprs)


def auc_trapz(x, y):
    order = np.argsort(x)
    x = x[order]
    y = y[order]
    return float(np.trapz(y, x))

fpr, tpr = roc_curve_manual(y_test, probs_test)
roc_auc = auc_trapz(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr)
plt.title(f"ROC Curve (AUC ≈ {roc_auc:.3f})")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

### Precision–Recall curve (manual)

```python
def pr_curve_manual(y_true, probs):
    y_true = np.asarray(y_true).reshape(-1)
    probs = np.asarray(probs).reshape(-1)

    thresholds = np.r_[np.sort(np.unique(probs))[::-1], -0.01]
    precisions = []
    recalls = []

    for t in thresholds:
        preds = (probs >= t).astype(int)
        tp, tn, fp, fn = confusion_counts(y_true, preds)
        precision, recall, f1, acc = classification_metrics(tp, tn, fp, fn)
        precisions.append(precision)
        recalls.append(recall)

    return np.array(recalls), np.array(precisions)

rec, prec = pr_curve_manual(y_test, probs_test)

plt.figure()
plt.plot(rec, prec)
plt.title("Precision–Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()
```

### Calibration plot (reliability diagram)

If the model says 0.80, do we actually see ~80% positives?

```python
def calibration_curve(y_true, probs, n_bins=10):
    y_true = np.asarray(y_true).reshape(-1)
    probs = np.asarray(probs).reshape(-1)

    bins = np.linspace(0, 1, n_bins + 1)
    bin_ids = np.digitize(probs, bins) - 1

    bin_true = []
    bin_pred = []

    for b in range(n_bins):
        mask = bin_ids == b
        if np.sum(mask) == 0:
            continue
        bin_true.append(np.mean(y_true[mask]))
        bin_pred.append(np.mean(probs[mask]))

    return np.array(bin_pred), np.array(bin_true)

bin_pred, bin_true = calibration_curve(y_test, probs_test, n_bins=10)

plt.figure()
plt.plot([0, 1], [0, 1])
plt.plot(bin_pred, bin_true, marker='o')
plt.title("Calibration Plot")
plt.xlabel("Mean predicted probability")
plt.ylabel("Fraction of positives")
plt.show()
```
